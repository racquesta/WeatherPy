# WeatherPy Analysis

```python
# Dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from citipy import citipy
import requests as req
import unidecode
import time
from math import sqrt
```

```python
# Build data frame of randomly generated lat and long
location_data = pd.DataFrame()
location_data['rand_lat'] = [np.random.uniform(-90,90) for x in range(1500)]
location_data['rand_lng'] = [np.random.uniform(-180, 180) for x in range(1500)]

# add closest city and country column
location_data['closest_city'] = ""
location_data['country'] = ""

#find and add closest city and country code
for index, row in location_data.iterrows():
    lat = row['rand_lat']
    lng = row['rand_lng']
    location_data.set_value(index, 'closest_city', citipy.nearest_city(lat, lng).city_name)
    location_data.set_value(index, 'country', citipy.nearest_city(lat, lng).country_code)
    
```

```python
location_data = location_data.drop_duplicates(['closest_city', 'country'])
location_data = location_data.dropna()
len(location_data['closest_city'].value_counts())
```

```python
# read in open weather map's country Id json
# downloaded from https://openweathermap.org/appid#work

api_city_data = pd.read_json('city.list.json')

for index, row in api_city_data.iterrows():
    lower_city = row['name'].lower()
    unaccented = unidecode.unidecode(lower_city)
    lower_country = row['country'].lower()
    api_city_data.set_value(index, 'name', unaccented)
    api_city_data.set_value(index, 'country', lower_country)
```

```python
api_city_data['id'].dtype
```

```python
#merge with random cities

location_data = location_data.rename(columns = {'closest_city': 'city'})
api_city_data = api_city_data.rename(columns = {'name': 'city'})

merged_df = location_data.merge(api_city_data, how = 'left', on = ('city', 'country'))
merged_df = merged_df.drop_duplicates(['city', 'country'])
len(merged_df)
```

```python
# id url example= 'api.openweathermap.org/data/2.5/weather?id=2172797'

# coord url example  = 'api.openweathermap.org/data/2.5/weather?lat=35&lon=139'

#city search use : api.openweathermap.org/data/2.5/weather?q=London,uk

# use &appid= for key

counter = 0 #for breaking and pausing
drop_list = [] # for cities without data
errors = 0
#create addition columns for open weather map data
cur_weather['date'] = ""
cur_weather['lat'] = ""
cur_weather['lng'] = ""
cur_weather['max_temp'] = ""
cur_weather['humidity'] = ""
cur_weather['clouds'] = ""
cur_weather['wind'] = ""

for index, row in cur_weather.iterrows():
    print('Now retrieving data for city #%s: %s, %s' % (index, row['city'], row['country']))
    if ((row['id']) == 0) and (row['coord'] != ""):
        lat = row['coord']['lat']
        lon = row['coord']['lon']
        url = 'https://api.openweathermap.org/data/2.5/weather?lat=%s&lon=%s&APPID=%s&units=imperial' % (lat, lon, wkey)  
    elif row['id'] != 0:
        loc_id = row['id']
        url = 'https://api.openweathermap.org/data/2.5/weather?id=%s&APPID=%s&units=imperial' % (loc_id, wkey)
    else:
        city = row['city']
        country = row['country']
        url = 'https://api.openweathermap.org/data/2.5/weather?q=%s,%s&APPID=%s&units=imperial' % (city, country, wkey)
    print(url)
    try:
        response = req.get(url).json()
        cur_weather.set_value(index, 'date', response['dt'])
        cur_weather.set_value(index, 'lat', response['coord']['lat'])
        cur_weather.set_value(index, 'lng', response['coord']['lon'])
        cur_weather.set_value(index, 'max_temp', response['main']['temp_max'])
        cur_weather.set_value(index, 'humidity', response['main']['humidity'])
        cur_weather.set_value(index, 'clouds', response['clouds']['all'])
        cur_weather.set_value(index, 'wind', response['wind']['speed'])
    except:
        print('Missing Info for city #%s: %s, %s' % (index, row['city'], row['country']))
        drop_list.append(index)
        errors += 1
    print('---------------------------------------------------------------------------')
    counter +=1
    if counter % 50 == 0:
        time.sleep(65)

print('---------------------------------------------------------------------------')
print('---------------------------------------------------------------------------')
print("Count: " + str(counter))
print("Errors: " + str(errors))
print('---------------------------------------------------------------------------')
print('---------------------------------------------------------------------------')
```

```python
#copy intact merged_df to new df
max_temp_df = merged_df
```

```python
# id url example= 'api.openweathermap.org/data/2.5/weather?id=2172797'

# coord url example  = 'api.openweathermap.org/data/2.5/weather?lat=35&lon=139'

#city search use : api.openweathermap.org/data/2.5/weather?q=London,uk

# use &appid= for key

counter = 0 #for breaking and pausing
drop_list = [] # for cities without data
errors = 0
#create addition columns for open weather map data
max_temp_df['date'] = ""
max_temp_df['lat'] = ""
max_temp_df['lng'] = ""
max_temp_df['max_temp'] = ""
max_temp_df['humidity'] = ""
max_temp_df['clouds'] = ""
max_temp_df['wind'] = ""


for index, row in max_temp_df.iterrows():
    print('Now retrieving forecast data for city #%s: %s, %s' % (index, row['city'], row['country']))
    if ((row['id']) == 0) and (row['coord'] != ""):
        lat = row['coord']['lat']
        lon = row['coord']['lon']
        url = 'https://api.openweathermap.org/data/2.5/forecast?lat=%s&lon=%s&APPID=%s&units=imperial' % (lat, lon, wkey)  
    elif row['id'] != 0:
        loc_id = row['id']
        url = 'https://api.openweathermap.org/data/2.5/forecast?id=%s&APPID=%s&units=imperial' % (loc_id, wkey)
    else:
        city = row['city']
        country = row['country']
        url = 'https://api.openweathermap.org/data/2.5/forecast?q=%s,%s&APPID=%s&units=imperial' % (city, country, wkey)
    print(url)
    try:
        response = req.get(url).json()
        response_path = response['list']
        temps = []
        for n in range(9): # a 24 hour period
            temps.append(response_path[n]['main']['temp_max'])
        max_index = temps.index(max(temps))
        max_temp_df.set_value(index, 'date', response_path[max_index]['dt'])
        max_temp_df.set_value(index, 'lat', response['city']['coord']['lat'])
        max_temp_df.set_value(index, 'lng', response['city']['coord']['lon'])
        max_temp_df.set_value(index, 'max_temp', response_path[max_index]['main']['temp_max'])
        max_temp_df.set_value(index, 'humidity', response_path[max_index]['main']['humidity'])
        max_temp_df.set_value(index, 'clouds', response_path[max_index]['clouds']['all'])
        max_temp_df.set_value(index, 'wind', response_path[max_index]['wind']['speed'])
        print(temps)
        print(max(temps))
        print(max_index)
        print(response_path[max_index]['main']['temp_max'])
    except:
        print('Missing Info for city #%s: %s, %s' % (index, row['city'], row['country']))
        drop_list.append(index)
        errors += 1
    print('---------------------------------------------------------------------------')
    counter +=1
#     if counter == 10:
#         break
    if counter % 50 == 0:
        time.sleep(65)


print('---------------------------------------------------------------------------')
print('---------------------------------------------------------------------------')
print("Count: " + str(counter))
print("Errors: " + str(errors))
print('---------------------------------------------------------------------------')
print('---------------------------------------------------------------------------')
```
